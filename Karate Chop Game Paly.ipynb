### **Let's See the Game Interface**

- **We have to click some specific area on the game screen in order to play game.**
- **We can take screenshort of whole game but it will take more time to process so we are taking the screenshort of some specific area**
- **Let's figure it out**

- **Inorder to click a specific area we have to find the cordinate of that area**

- **We will use the ImageGrab.grab() function to take a screen short!**

- **To take screenshort of a specific area or location in your screen we have to give 4 parameters top_left_x, top_left_y, bottom_righr_x, bottom_right_y, so let's find our cordinate**

- To know the exact coordinates, take the screenshort of your screen and open with paint, then click on the view bar and select Rulers checkbox! Then your pointer will show the coordinates below at bottom left!

```python
# Take the screenshort of Play Button Location
# x1, y1, x2, y2 Co-Ordinates
# These values depends upon the Screen Resolutions and the position of the game on screen. 
# I am taking these value according to my screen, it may differ on other PC

x1 = 225
y1 = 710
x2 = x1 + 120
y2 = y1 + 55

img = ImageGrab.grab(bbox=(x1, y1, x2, y2))  # Take ScreenShort and store in img

ssc = np.array(img)                          # Convetr image to numpy array to plot using matplotlib
plt.imshow(ssc)                              # Show image on screen
plt.show()
```

```python
# Take the screenshort of Left Area Location
# x1, y1, x2, y2 Co-Ordinates
# These values depends upon the Screen Resolutions and the position of the game on screen. 
# I am taking these value according to my screen, it may differ on other PC

left_x1 = 130
left_y1 = 420
left_x2 = left_x1 + 140
left_y2 = left_y1 + 200

left_cordinate = (left_x1, left_y1, left_x2, left_y2) # we are storing these cordinates to a single variable to use later

img = ImageGrab.grab(bbox= left_cordinate)            # Take ScreenShort and store in img  

ssc = np.array(img)                                   # Convetr image to numpy array to plot using matplotlib
plt.imshow(ssc)                                       # Show image on screen
plt.show()
```

```python
# Take the screenshort of Right Area Location

right_x1 = 370
right_y1 = 420
right_x2 = right_x1 + 140
right_y2 = right_y1 + 200

right_cordinate = (right_x1, right_y1, right_x2, right_y2)

img = ImageGrab.grab(bbox=right_cordinate)

ssc = np.array(img)
plt.imshow(ssc)
plt.show()
```

**Now Let's make a function which will return the screensort of the specific area/location!**

```python
def grab_screen(bbox=None):
    screenshort = ImageGrab.grab(bbox)
    screenshort = np.array(screenshort)
    # since we have to use this image in the opencv that's why we have to convert the RGB image into BGR image!
    screenshort = cv2.cvtColor(screenshort, cv2.COLOR_RGB2BGR)
    return screenshort
```

```python
# The number of seconds to pause after every keypress event.
pyautogui.PAUSE = 0
```

**Now, Let's add all the objects present in the Game**

```python
wood_left = cv2.imread('woodleft.jpg')
wood_right = cv2.imread('woodright.jpg')
```

**Height and Weight of the wood image**

```python
w = wood_left.shape[1]
h = wood_left.shape[0]
```

## **Now Let's integrate all to play the game!**

```python
import time
import keyboard
```

```python
# At the start of the game, the player position in on left so we are tracking the movement of the player
left = True

# Left area click cordinate to play game
click_x = 225
click_y = 710


while True:
    # Find if the player is on left or right site in game
    if left:
        # If the player is on left site then take Screenshort of left area
        img = grab_screen(bbox = left_cordinate)
        
        # For left area your enemy is left Wood 
        wood = wood_left
    else:
        # If the player is on right site then take Screenshort of right area
        img = grab_screen(bbox = right_cordinate)
        
        # For right area your enemy is right Wood
        wood = wood_right

        
    # Cut off alpha
    img_remove = img[:,:,:3]

    # Detect enemy wood
    result = cv2.matchTemplate(img_remove, wood, cv2.TM_CCOEFF_NORMED)
    
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
    img = img.copy()
    
    # Probability to detect eenmy in image
    if max_val > .80:
        left = not left
        if left:
            # Cordinate to click on left part
            click_x = 220
        else:
            # Cordinate to click on right part
            click_x = 345
        
        # Draw rectangle on enmey
        cv2.rectangle(img, max_loc, (max_loc[0] + w, max_loc[1] + h), (0,0,255), 3)

    # Show the result image
    cv2.imshow('Karate Chop Kick', img)
    
    #click on screen
    pyautogui.click(x = click_x, y = click_y)
    
    # you can use sleep to slow down these process
#     time.sleep(0.01)
    if keyboard.is_pressed('q') or cv2.waitKey(1) == 27:
        break
#     if cv2.waitKey(1) == 27:
#         break
        
    
cv2.destroyAllWindows()
```
